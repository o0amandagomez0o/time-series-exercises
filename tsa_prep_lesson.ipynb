```python
#change dtype to datetime
df.sale_date = pd.to_datetime(df.sale_date)
```

```python
#set index to date
df = df.set_index('sale_date').sort_index()
```

### one observation: the combination of store, item, and date

- how could we transform the data such that each row represents a unique date?

    - A: aggregate or filter

```python
df.resample('D').sale_amount.sum()
```

### the way we choose to aggregate or data is going to depend on how we approach the problem we are trying to solve

- lets say we want to look at sales (sum sale_amount) for *each* store for *each* day

```python
df.groupby('store_id').resample('D').sale_amount.sum()
```

```python
df.groupby('store_id').resample('D').sale_amount.sum().unstack(0)
```

```python
df.pivot_table('sale_amount', index='sale_date', columns='store_id', aggfunc='sum')
```

### pick one of either the .groupby().resample()...unstack() OR .pivot_table
